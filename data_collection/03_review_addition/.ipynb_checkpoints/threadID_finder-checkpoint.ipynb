```python
import re
```

```python
path = '/Users/j.kanlaya/Downloads/Mike/data_collection/03_review_addition/'
```

```python
text = open(path + "02+03_341dataset.txt", "r").read()
```

```python
# text
```

```python
id_list = []
# match = re.search(r'<h1>\d\d\d\d\d\d\d\d</h1>', text)
match01 = re.findall(r'<h1>\d+</h1>', text)
for thread_id in match01:
    thread_id = thread_id.split("<h1>")
    thread_id = thread_id[1].split("</h1>")
    id_list.append(thread_id[0])

match02 = re.findall(r'<h1>\d+<h1>', text)
for thread_id in match02:
    thread_id = thread_id.split("<h1>")
    id_list.append(thread_id[1])
id_list
```

```python
len(id_list)
```

```python
f=open('threadID_list.txt','w')
for element in id_list:
    f.write(element+'\n')

f.close()
```
