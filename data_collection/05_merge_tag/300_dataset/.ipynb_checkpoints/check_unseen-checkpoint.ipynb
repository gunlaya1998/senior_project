```python
import csv
import pandas as pd
```

```python
train_df = pd.read_csv('300_loc_data.csv')
train_df
```

```python
pred_df = pd.read_csv('unseen_LocList.csv')
pred_df
```

```python
pred = list(pred_df['Word'])
pred = list(dict.fromkeys(pred))
pred
```

```python
train = list(train_df['Word'])
train = list(dict.fromkeys(train))
train
```

```python
unseen = []
seen = []
for word in pred:
    if(word not in train):
        unseen.append(word)
    else:
        seen.append(word)
```
