```python
import pandas as pd
import numpy as np
import csv
```

## Read CSV

```python
act_df = pd.read_csv('300_loc_data.csv')
act_df
```

```python
pred_df = pd.read_csv('300_loc_predict.csv')
pred_df
```

## Create Actual location dictionary with thread_no as key

```python
act_dict = {}
act_count = 0
for i in range(len(act_df)):
    if (act_df.iloc[i][0] in act_dict):
        if(act_df.iloc[i][2] not in act_dict[act_df.iloc[i][0]]):
            act_dict[act_df.iloc[i][0]].append(act_df.iloc[i][2])
            act_count += 1   
    else:
        act_dict[act_df.iloc[i][0]] = [act_df.iloc[i][2]]
        act_count += 1
act_dict
```

## Create Predicted location dictionary with thread_no as key

```python
pred_dict = {}
pred_count = 0
for i in range(len(pred_df)):
    if pred_df.iloc[i][0] in pred_dict:
        if(pred_df.iloc[i][2] not in pred_dict[pred_df.iloc[i][0]]):
            pred_dict[pred_df.iloc[i][0]].append(pred_df.iloc[i][2])
            pred_count += 1
    else:
        pred_dict[pred_df.iloc[i][0]] = [pred_df.iloc[i][2]]
        pred_count += 1
pred_dict
```

# Count predicted_loc occurence in actual_loc

```python
len(pred_dict)
```

```python
thread_score_list = []
total_score = 0
for key in pred_dict:
    thread_score = 0
    if(key in act_dict):
        for loc in pred_dict[key]:
            if(loc in act_dict[key]):
                thread_score += 1
                total_score += 1
        thread_score = thread_score / len(act_dict[key])
    thread_score_list.append((key, round(thread_score, 4)))
```
