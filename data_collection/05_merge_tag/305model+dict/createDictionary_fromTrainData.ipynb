```python
import pandas as pd
import numpy as np
import csv
import json
import re
import string
```

```python
df = pd.read_csv('305_loc_data.csv')
df
```

```python
loc = df['Word'].unique()
loc
```

```python
with open('TrueLoc_dic.txt', 'w', encoding='utf8') as fp:
    fp.write('\n'.join(json.dumps(i, ensure_ascii=False) for i in loc) +'\n')
```

```python
f = open('TrueLoc_dic.txt', 'r', encoding="utf8").read()
print(f)
```

```python
text = re.sub(r"\"", "", f, flags=re.I)
print(text)
```
