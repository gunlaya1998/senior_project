## Create Actual location dictionary with thread_no as key

```python
act_dict = {}
act_count = 0
for i in range(len(act_df)):
    tid = act_df.iloc[i][0]
    word = act_df.iloc[i][3]
    if (tid in act_dict):
        if(word not in act_dict[tid]):
            act_dict[tid].update( {word: 1} )
        else:
            act_dict[tid][word] += 1
    else:
        act_dict[tid] = {word: 1}
    act_count += 1
act_dict
```

## Create Predicted location dictionary with thread_no as key

```python
pred_dict = {}
pred_count = 0
for i in range(len(pred_df)):
    tid = pred_df.iloc[i][0]
    word = pred_df.iloc[i][3]
    if tid in pred_dict:
        if(word not in pred_dict[tid]):
            pred_dict[tid].update( {word: 1} )
        else:
            pred_dict[tid][word] += 1
    else:
        pred_dict[tid] = {word: 1}
    pred_count += 1
pred_dict
```

# Count predicted_loc occurence in actual_loc

```python
thread_act = {}
for tid in act_dict.keys():
    act = 0
    for loc in act_dict[tid]:
        act += act_dict[tid][loc]
    thread_act[tid] = act
thread_act
```

```python
thread_pred = {}
pred_check = 0
for tid in pred_dict.keys():
    pred = 0
    for loc in pred_dict[tid]:
        pred += pred_dict[tid][loc]
        pred_check += pred_dict[tid][loc]
    thread_pred[tid] = pred
thread_pred
```

```python
pred_check
```

```python
thread_score_list = []
total_score = 0
pred_check=0
for tid in pred_dict:
    thread_score = 0
    if(tid in act_dict):
        for loc in pred_dict[tid]:
            if(loc in act_dict[tid]):
                if(pred_dict[tid][loc] <= act_dict[tid][loc]):
                    thread_score += pred_dict[tid][loc]
                    total_score += pred_dict[tid][loc]
                else:
                    thread_score += act_dict[tid][loc]
                    total_score += act_dict[tid][loc]
        thread_score = thread_score / thread_act[tid]
    
    thread_score_list.append((tid, round(thread_score, 4)))
```
