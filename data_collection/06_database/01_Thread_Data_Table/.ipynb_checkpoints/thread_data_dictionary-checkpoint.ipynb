```python
pred_dict = {}
dup_filter = {}
for i in range(len(pred_df)):
    tid = int(pred_df.iloc[i]['Thread_id'])
    pname = str(pred_df.iloc[i]['Word'])
    lat_data = float(pred_df.iloc[i]['LAT'])
    lon_data = float(pred_df.iloc[i]['LON'])
    
    if tid in pred_dict.keys() and pd.notna(lat_data):
        if(pname not in dup_filter[tid]):
            pred_dict[tid].append({"placeId":i+1, "placeName":pname, "lat":lat_data, "lon":lon_data})
            dup_filter[tid].append(pname)
            
    elif pd.notna(lat_data):
        data_list = [{"placeId":i+1, "placeName":pname, "lat":lat_data, "lon":lon_data}]
        pred_dict[tid] = data_list
        
        dup_filter[tid] = [pname]
        
pred_dict
```
