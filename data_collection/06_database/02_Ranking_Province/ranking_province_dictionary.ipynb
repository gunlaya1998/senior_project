```python
with open('provinceId_count.json') as json_file: 
    pdata = json.load(json_file)
pdata
```

```python
pdict = {}
for tid in pdata.keys():
    for pid in pdata[tid]:
        if pid in pdict:
            pdict[pid].append(tid)
        else:
            pdict[pid] = [tid]

pdict
```

```python
with open('/Users/j.kanlaya/Downloads/Mike/data_collection/06_database/01_Thread_Data_Table/trendscore_data.json') as json_file: 
    score_dict = json.load(json_file)
score_dict
```

```python
pscore_dict = {}
for pid in pdict.keys():
    try:
        score = 0
        for tid in pdict[pid]:
            score += score_dict[tid]
        pscore_dict[pid] = score
    
    except:
        pscore_dict[pid] = 0

pscore_dict
```

```python
sorted_dict = dict(sorted(pscore_dict.items(), key=lambda item: item[1], reverse=True))
sorted_dict
```

```python
rank = list(sorted_dict.keys())[:]
print(rank)
```
