```python
threadId_file = open("threadID_list.txt", "r")
threadId_list = []
while True: 
    line = threadId_file.readline() 
    line = line.strip('\n')
    threadId_list.append(line)
    if not line: 
        break
threadId_list = threadId_list[:-1]
# 38190914
threadId_list
```

```python
def checkIfDuplicates_1(listOfElems):
    ''' Check if given list contains any duplicates '''
    if len(listOfElems) == len(set(listOfElems)):
        return False
    else:
        return True
```

```python
result = checkIfDuplicates_1(threadId_list)
if result:
    print('Yes, list contains duplicates')
else:
    print('No duplicates found in list')    
```

```python
text = open("300_byLine.txt", 'r') 
# limit = 0
text_list = []
# imgIndex_dict = {}
thread_index = 0
while True: 
#     limit += 1
    line = text.readline() 
#     if not line:
    if(line == ""):
        break
        
    tokens = word_tokenize(line, engine="newmm")
    
    tag_count = 0
    word_index = 0
#     thread_index = 0
    tag_index = []
    while (word_index < len(tokens)):
        if("classimg-in-post" in tokens[word_index]):
            tag_index.append(word_index - tag_count - 3)
            tag_count += 1
            word_index += 1
            
            while ("/>" not in tokens[word_index]):
                tag_count += 1
                word_index += 1
                if(word_index == len(tokens)):
                    break
                
        else:
            word_index += 1
            
#     if(tag_count == 0):    
#         thread_index += 1
        
#     thread_index += 1
#     imgIndex_dict[threadId_list[thread_index]] = tag_index
    tmp = []
    tmp.append(threadId_list[thread_index])
    tmp.append(tag_index)
    
    tmp = tuple(tmp)
    text_list.append(tmp)
    thread_index += 1
    
# print(imgIndex_dict)
print(len(text_list))
print(text_list)
```
